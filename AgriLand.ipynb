```python
def landclass(c):
    if c['land'] == 0:
        return 'ไม่มีที่ดิน'
    elif ((c['land'] > 0) & (c['land'] < 800)):
          return 'น้อยกว่า 2 ไร่' 
    elif ((c['land'] >= 800) & (c['land'] < 2000)):
          return '2 - 4 ไร่'
    elif ((c['land'] >= 2000) & (c['land'] < 4000)):
          return '5 - 9 ไร่'
    elif ((c['land'] >= 4000) & (c['land'] < 8000)):
          return '10 - 19 ไร่'
    elif ((c['land'] >= 8000) & (c['land'] < 16000)):
          return '20 - 39 ไร่'
    elif ((c['land'] >= 16000)):
        return 'ตั้งแต่ 40 ไร่ขึ้นไป'

ses2011['landclass'] = ses2011.apply(landclass, axis=1)
ses2013['landclass'] = ses2013.apply(landclass, axis=1)
ses2015['landclass'] = ses2015.apply(landclass, axis=1)
ses2017['landclass'] = ses2017.apply(landclass, axis=1)
ses2019['landclass'] = ses2019.apply(landclass, axis=1)
```

## Decile

```python
import numpy as np
import pandas as pd
from pandas._libs.lib import is_integer

def weighted_qcut(values, weights, q, **kwargs):
    'Return weighted quantile cuts from a given series, values.'
    if is_integer(q):
        quantiles = np.linspace(0, 1, q + 1)
    else:
        quantiles = q
    order = weights.iloc[values.argsort()].cumsum()
    bins = pd.cut(order / order.iloc[-1], quantiles, **kwargs)
    return bins.sort_index()
```

```python
ses2011 = ses2011[ses2011.landown != 0]
ses2013 = ses2013[ses2013.landown != 0]
ses2015 = ses2015[ses2015.landown != 0]
ses2017 = ses2017[ses2017.landown != 0]
ses2019 = ses2019[ses2019.landown != 0]
```

```python
ses2011['weighted_qcut'] = weighted_qcut(ses2011['landown'], ses2011['A52'], 10, labels=False)
ses2013['weighted_qcut'] = weighted_qcut(ses2013['landown'], ses2013['A52'], 10, labels=False)
ses2015['weighted_qcut'] = weighted_qcut(ses2015['landown'], ses2015['A52'], 10, labels=False)
ses2017['weighted_qcut'] = weighted_qcut(ses2017['landown'], ses2017['A52'], 10, labels=False)
ses2019['weighted_qcut'] = weighted_qcut(ses2019['landown'], ses2019['A52'], 10, labels=False)
```

```python
decileland2011 = ses2011.groupby(['weighted_qcut'])[['A52', 'landown']].sum()
decileland2013 = ses2013.groupby(['weighted_qcut'])[['A52', 'landown']].sum()
decileland2015 = ses2015.groupby(['weighted_qcut'])[['A52', 'landown']].sum()
decileland2017 = ses2017.groupby(['weighted_qcut'])[['A52', 'landown']].sum()
decileland2019 = ses2019.groupby(['weighted_qcut'])[['A52', 'landown']].sum()
decileland2019
```

```python
import xlsxwriter
writer = pd.ExcelWriter('Decilebyland.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
decileland2011.to_excel(writer, sheet_name='2011')
decileland2013.to_excel(writer, sheet_name='2013')
decileland2015.to_excel(writer, sheet_name='2015')
decileland2017.to_excel(writer, sheet_name='2017')
decileland2019.to_excel(writer, sheet_name='2019')
# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
